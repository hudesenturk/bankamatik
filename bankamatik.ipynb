```python
# bankamatik
# 29.04.2023

sifre="huhu"

sifre_kontrol=0

hesaplar=[1500,1000]

menu=""" *** ZIRA BANK ***

1- Bakiye sorma
2- Para çekme
3- Para yatırma
4- Hesaplar arası transfer
5- Hesap ekleme
6- Hesap silme
7- Çıkış"""

for i in range(3):
    pw=input("Lütfen şifrenizi giriniz")
    if pw==sifre:
        sifre_kontrol=1
        break


if sifre_kontrol!=1:
    print("Şifrenizi girmediniz:")
else:
    print("ZIRA BANK'A HOŞ GELDİNİZ.")

    while True:
        print(menu)
        secim=input("Seçiminiz: ")

        if secim=="1":
            for i in range(len(hesaplar)):
                print(i+1,". hesabınız: ",hesaplar[i])

        if secim=="2":

            for i in range(len(hesaplar)):
                print(i+1,". hesabınız: ",hesaplar[i])

            hesap=int(input("Toplam "+str(len(hesaplar))+" hesabınız var, hangisinden para çekmek istiyorsunuz?"))



            if hesap>len(hesaplar):
                print("Yanlış hesap numarası girdiniz")
            else:
                miktar_cek=int(input("Ne kadar çekeceksiniz: "))

                if miktar_cek > hesaplar[hesap-1]:
                    print("Bakiye yetersiz")
                else:
                    hesaplar[hesap-1]=hesaplar[hesap-1]-miktar_cek

        if secim=="3":
            hesap1=int(input("Toplam "+str(len(hesaplar))+" hesabınız var, hangisine para yatırmak istiyorsunuz?"))

            if hesap1 > len(hesaplar):
                print("Yanlış hesap numarası girdiniz.")
            else:
                miktar_yatir=int(input("Ne kadar yatıracaksınız: "))
                hesaplar[hesap1-1]+=miktar_yatir


        if secim=="5":
            ekle_miktar=int(input("Yeni hesabınızı kaç lira ile açmak istersiniz: "))
            hesaplar.append(ekle_miktar)


        if secim=="6":
            for i in range(len(hesaplar)):
                print(i+1,". hesabınız",hesaplar[i])

            hesap_sil=int(input("Toplam "+str(len(hesaplar))+ "hesabınız var, hangisini silmek istiyorsunuz: "))

            aktar=int(input("Bu hesabınızda toplam "+str(hesaplar[hesap_sil-1])+" TL var, hangi hesaba transfer etmek istersiniz"))

            hesaplar[aktar-1]+=hesaplar[hesap_sil - 1]
            hesaplar.pop(hesap_sil-1)



        elif secim=="7":
            print("Program sona erdi.")
            break


```
